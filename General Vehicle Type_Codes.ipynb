## Data Visualization

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['State'], df['total'])
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Total vehicles')
plt.title('Total Number of Vehicles in Each State')
plt.show()
           
```

```python
import matplotlib.pyplot as plt

# create a pie chart of Total Vehicles by Type
labels = ['L1_L2', 'L2', 'L2_250W', 'L5_slow', 'L5', 'M1','B1']
sizes = [
    df['L1_L2'].sum(),
    df['L2'].sum(),
    df['L2_250W'].sum(),
    df['L5_slow'].sum(),
    df['L5'].sum(),
    df['M1'].sum(),
    df['B1'].sum()
]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'orange', 'pink', 'violet']
explode = (0, 0, 0.1, 0, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.legend(labels, loc="best", bbox_to_anchor=(1, 0.25, 0.5, 0.5))
plt.axis('equal')
plt.title('Percentage of Each Vehicle Category in the Country')
plt.show()
```

```python
import seaborn as sns

# select columns to include in the heatmap
cols = ['State', 'L1_L2', 'L2', 'L2_250W', 'L5_slow', 'L5', 'M1', 'B1']

# create a new dataframe with the selected columns
df_heatmap = df[cols]

# create the heatmap
sns.heatmap(df_heatmap.corr(), cmap='coolwarm', annot=True)

# set the title
plt.title('Heatmap')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# create a list of columns to include in the histogram
columns = ['L1_L2', 'L2', 'L2_250W']

# create subplots with 1 row and 3 columns
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# loop through each column and create a histogram
for i, col in enumerate(columns):
    # create the histogram
    axs[i].hist(df[col], bins=10)
    
    # set the title and x/y labels
    axs[i].set_title(col)
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

# add an overall title to the plot
fig.suptitle('Two Wheelers')

# adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# show the plot
plt.show()
```
